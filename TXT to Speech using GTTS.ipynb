```python
import pyttsx3
from IPython.display import Audio

text = ''' Hello my sister name is Vinuthna , she is studying 10 th class and now her CBSE board exams are started and 
           she have science test on day after tomorrow but i donot thing that she completed her revision.I hope she passes the exam'''

audio = pyttsx3.init()
audio.setProperty('rate' , 150)
audio.setProperty('volume' , 0.8)

#change the voice

voice = audio.getProperty('voices')

#0 for male voice and 1 for female voice

audio.setProperty('voice' , voice[0].id) #for male voice
#audio.setProperty('voice' , voice[1].id) #for female voice

#text to speech conversion
audio.say(text)

#save audio file
audio.save_to_file(text, 'test_male_voice.mp3')

audio.runAndWait()
```

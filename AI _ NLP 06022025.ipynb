# Word2vec

```python
import gensim
from gensim.models import Word2Vec

#sample sentences for training the model
sentence = [
    ["my" , "name" , "is" , "Ramya"],
    ["I" , "love" , "watching" , "dramas"],
    ["I" , "want" , "to" , "travel" , "south korea"],
]

#training the word2vec model
model = Word2Vec(sentence, vector_size = 100 , window = 5 , min_count = 1 , sg = 1)

#finding the vector with target word
vector = model.wv['Ramya']
print("vector for 'Ramya' : ", vector)

#finding similar words
similar_words = model.wv.most_similar('travel' , topn = 1)
print("words similar to 'travel':" , similar_words)
```

```python
import gensim
from gensim.models import Word2Vec

#dataset
sentences = [
    ["I" , "am" , "currently" , "learning" , "full stack " , "datascience " , "and " , "AI"],
    ["from" , "Naresh senapati" , "of" , "Naresh IT"],
    ["I" , "need" , "to" , "apply" , "for" , "jobs"],
]

#training the dataset
model = Word2Vec(sentences , min_count = 1 , sg = 1 , vector_size = 100 , window = 5)

#finding the vector of target word
vector = model.wv['currently']
print("vector for 'currently' : " , vector)

#finding similar words
similar_word = model.wv.most_similar('learning')
print("most similar to 'learning' : " , similar_word)
```

```python
import gensim
from gensim.models import Word2Vec

#sample sentences for training the model
sentence = [
    ["my" , "name" , "is" , "Ramya"],
    ["I" , "love" , "watching" , "dramas"],
    ["I" , "want" , "to" , "travel" , "south korea"],
    ["I" , "love" , "eating " , "good" , "food" ],
]

#training the dataset
cbow_model = Word2Vec(sentence , window = 2 , min_count = 1 , vector_size = 100 , sg = 0 )

skipgram_model = Word2Vec(sentence , window = 2 , min_count = 1 , vector_size = 100 , sg = 1 )

#getting vector for the target word
target_word = "Ramya"
cbow_vector = cbow_model.wv[target_word]
skipgram_vector = skipgram_model.wv[target_word]

print(f"Cbow vector for '{target_word}' : " , cbow_vector)
print(f"skipgram vector for '{target_word}' : " , skipgram_vector )

#finding similiar words

cbow_similar_words = cbow_model.wv.most_similar(target_word, topn = 2)
skipgram_similar_words = skipgram_model.wv.most_similar(target_word , topn = 2)

print(f"cbow similar_words to '{target_word}':" , cbow_similar_words)
print(f"skipgram similar_words to '{target_word}':" , skipgram_similar_words )



                      
```
